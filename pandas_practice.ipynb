## 基本pandas操作  
.read_csv 讀取csv檔案  
.discribe() 初步了解資料表  
.isna 檢查是否為na值  
.info()出布料解資料表  
.concat :dataframe合併  
.to_csv 儲存成csv  
.rename  
.merge 依照不同的dataframe有不同的key去做merge  

```python
# 載入pandas模組
import pandas as pd

# 載入資料
data = pd.read_csv(r"./2021.csv")
# 印出載入的資料
#print(data)
#print(data.describe())
#print(data.isna())
print(data.info())
```

```python
# 準備需要合併的資料總表
data_mother = pd.read_csv(r"./飲料店總表0302.csv")
print(data_mother)
```

```python
# 過濾data做篩選欄位(剩下台北)
data_of_tpe = data[data["縣市"] == "台北市"]
#print(data_of_tpe)


# 過濾data做篩選欄位
data_of_tpe_new = data[data["縣市"] == "新北市"]
#print(data_of_tpe_new)


# 把過濾過的資料做合併成想要的資料
merge_old_new_tpe = pd.concat([data_of_tpe, data_of_tpe_new], ignore_index=True)
print(merge_old_new_tpe)
```

```python
# 合併後並儲存 成csv
merge_old_new_tpe.to_csv("old_and_new_taipei.csv", index=False, encoding="utf-8-sig")
```

```python
# 資料篩選
filter_data = data[data["納稅單位"]==287000]
print(filter_data) 
```

## 使用numpy對資料表對數值運算

```python
# 載入pandas模組
import pandas as pd
import numpy as np
# 載入資料
data = pd.read_csv(r"./2021.csv")

print(data)

```

## 把變數依照欄位相同的做concat
pd.merge 指定欄位merge 依據相同名字  
可指定left right join

```python
# .rename把要合併的column改成相同名字  村里改成neughborhood
merge_old_new_tpe = merge_old_new_tpe.rename(columns={'村里':'neighborhood'})
print(merge_old_new_tpe)
```

```python
# 使用.merge合併data 指定合併依據的column和想要合併的欄位 left join 保留左邊的完整data
merge_data_mother_all = pd.merge(data_mother,merge_old_new_tpe[['neighborhood','中位數']], on='neighborhood' , how= 'left')
print(merge_data_mother_all)
```

```python
# 儲存成csv
merge_data_mother_all.to_csv(
    "儲存合併里和中位數.csv", index=False, encoding="utf-8-sig"
)
```
